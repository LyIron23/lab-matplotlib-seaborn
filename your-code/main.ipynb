#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1

plt.plot(x, y, label='y = 2x', linestyle='-', color='blue')
plt.plot(x, z, label='z = x^2', linestyle='--', color='red')
plt.xlabel('x')
plt.ylabel('y and z')
plt.title('Plot of y = 2x and z = x^2')
plt.legend()

plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
fig, axs = plt.subplots(nrows=1, ncols=2)

# Plot y = 2x on the first subplot
axs[0].plot(x, y, linestyle='-', color='blue')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].set_title('Plot of y = 2x')

# Plot z = x^2 on the second subplot
axs[1].plot(x, z, linestyle='--', color='red')
axs[1].set_xlabel('x')
axs[1].set_ylabel('z')
axs[1].set_title('Plot of z = x^2')

plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# first subplot
axs[0].plot(x, y, linestyle='-', color='blue')
axs[0].set_xlabel('x', fontsize=12, fontweight='bold')  
axs[0].set_ylabel('y', fontsize=12, fontweight='bold')
axs[0].set_title('Plot of y = 2x', fontsize=14, fontweight='bold')  

# second subplot
axs[1].plot(x, z, linestyle='--', color='red')
axs[1].set_xlabel('x', fontsize=12, fontweight='bold') 
axs[1].set_ylabel('z', fontsize=12, fontweight='bold')  
axs[1].set_title('Plot of z = x^2', fontsize=14, fontweight='bold') 

plt.show()
```

#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
y1 = x ** 2
y2 = np.exp(x)

# a figure and axis object
fig, ax = plt.subplots(figsize=(8, 6))

# y = x^2 
ax.plot(x, y1, label='y = x^2', color='blue')

# y = e^x 
ax.plot(x, y2, label='y = e^x', color='red')

# axis labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Plot of y = x^2 and y = e^x (Normal Scale)')

# put x and y scales to logarithmic
ax.set_xscale('log')
ax.set_yscale('log')

ax.legend()

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
import matplotlib.pyplot as plt
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
median_steps_by_month_encoded = fitbit.groupby('Months_encoded')['Steps'].median()
months_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.figure(figsize=(12, 4))
median_steps_by_month_encoded.plot(kind='bar', color='blue')
plt.title('Median Steps by Month (Encoded)')
plt.xlabel('Month (Encoded)')
plt.ylabel('Median Steps')
plt.xticks(range(12), months_labels)  # Use custom month labels

# Create a folder called 'figures' 
import os
if not os.path.exists('figures'):
    os.makedirs('figures')

figure_path = os.path.join('figures', 'median_steps_by_month_encoded.png')
plt.savefig(figure_path)

# Show the plot
plt.show()
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
median_steps_by_day_type = fitbit.groupby('Work_or_Weekend')['Steps'].median()

plt.figure(figsize=(12, 4))
median_steps_by_day_type.plot(kind='bar', color='green')
plt.title('Median Steps by Work or Weekend')
plt.xlabel('Day Type')
plt.ylabel('Median Steps')
plt.xticks([0, 1], ['Weekend', 'Workday'])

figure_path = os.path.join('figures', 'median_steps_by_work_or_weekend.png')
plt.savefig(figure_path)

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
feature_combinations = [('Minutes Lightly Active', 'Steps'),('Minutes Very Active', 'Steps'),('Minutes Sedentary', 'Steps')]

fig, axs = plt.subplots(1, 3)

# Loop through the feature combinations
for i, (feature1, feature2) in enumerate(feature_combinations):
    plt.sca(axs[i])
    plt.scatter(fitbit[feature1], fitbit[feature2])
    plt.title(f'{feature1} vs {feature2}')
    plt.xlabel(feature1)
    plt.ylabel(feature2)

# Save the figure
figure_path = os.path.join('figures', 'scatter_plots.png')
plt.savefig(figure_path)

plt.show()
```

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it.
